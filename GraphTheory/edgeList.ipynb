### Library


```python
import networkx as nx
import pandas as pd
```

### Edge List


```python
df = pd.read_csv('edgeList.csv', header=None)
df.columns = ["source", "target", "weight"]

G = nx.from_pandas_edgelist(df, edge_attr=True, create_using=nx.MultiDiGraph())
print(df)
```

### Adj Matrix


```python
nx.to_pandas_adjacency(G, dtype=int)
```
